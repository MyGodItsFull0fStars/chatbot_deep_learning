```python
import json
import json_utils


def change_values(dir_name: str, file_name: str = json_utils.DEFAULT_FILE_NAME):
    json_data = None

    if dir_name is None or len(dir_name) == 0:
        accuracy_loss_file_name = file_name
    else:
        dir_name = dir_name.removesuffix('/')
        accuracy_loss_file_name = f'{dir_name}/{file_name}'

    with open(accuracy_loss_file_name, 'r') as read_file:
        json_data = json.load(read_file)

    json_data['loss'] = [data for data in json_data['loss']
                         if 'average' not in data[0]]

    with open(accuracy_loss_file_name, 'w') as out_file:
        json.dump(json_data, out_file)

for dir_name in ['models_1.5_hl', 'models_1_hl', 'models_no_hl']:
    change_values(dir_name)
```
