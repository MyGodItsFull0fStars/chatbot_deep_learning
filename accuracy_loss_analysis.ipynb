# Imports

```python
from json_utils import AccuracyLossData
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
from typing import List, Tuple


def multi_plot(accuracy_loss_data: AccuracyLossData):
    fig, (ax_accuracy, ax_loss) = plt.subplots(1, 2)
    fig.suptitle(accuracy_loss_data.model_name)
    fig.set_figheight(10)
    fig.set_figwidth(25)

    ax_accuracy.plot(accuracy_loss_data.accuracies.values())
    ax_accuracy.set_title('Accuracy')

    ax_loss.plot(accuracy_loss_data.losses.values())
    ax_loss.set_title('Loss')
```

```python

accuracy_loss_1_5_hl = AccuracyLossData('models_1.5_hl')
multi_plot(accuracy_loss_1_5_hl)
```

```python
accuracy_loss_1_hl = AccuracyLossData('models_1_hl')
multi_plot(accuracy_loss_1_hl)
```

```python
accuracy_loss_no_hl = AccuracyLossData('models_no_hl')
multi_plot(accuracy_loss_no_hl)
```

```python
def compare_multiple_data(accuracy_loss_data_list: List[Tuple[str, AccuracyLossData]]):

    fig, (ax_accuracy, ax_loss) = plt.subplots(1, 2)
    fig.suptitle('Comparison of Model Data')
    fig.set_figheight(10)
    fig.set_figwidth(25)

    ax_accuracy.set_title('Accuracy')
    for line_name, accuracy_loss_data in accuracy_loss_data_list:
        ax_accuracy.plot(accuracy_loss_data.accuracies.values(), label=line_name)
        ax_loss.plot(accuracy_loss_data.losses.values())

    fig.legend()


accuracy_loss_list: List[Tuple[str, AccuracyLossData]] = [
    ('Model 1.5 HiddenLayer', accuracy_loss_1_5_hl), 
    ('Model 1 HiddenLayer', accuracy_loss_1_hl), 
    ('Model no HiddenLayer', accuracy_loss_no_hl)
    ]

compare_multiple_data(accuracy_loss_list)
```
